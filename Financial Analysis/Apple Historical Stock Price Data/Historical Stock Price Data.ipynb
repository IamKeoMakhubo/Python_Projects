```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
symbol ="AAPL"
start_date = "2010-01-01"
end_date = "2023-01-01"
```

```python
stock_data = yf.download(symbol, start=start_date, end=end_date)
```

```python
print(stock_data.head())
```

```python
stock_data['Daily Return'] = stock_data['Adj Close'].pct_change()
```

```python
plt.figure(figsize=(12, 6))
```

```python
plt.subplot(2, 1, 1)
plt.plot(stock_data['Adj Close'], label='Adj Close Price', color='maroon')
plt.title(f'{symbol} Stock Price Analysis')
plt.ylabel('Stock Price')
plt.legend()
```

```python
plt.subplot(2, 1, 2)
plt.plot(stock_data['Daily Return'], label='Daily Returns', color='red')
plt.title(f'{symbol} Daily Returns Analysis')
plt.xlabel('Date')
plt.ylabel('Daily Returns')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
avg_daily_return = stock_data['Daily Return'].mean()
print(f"Average Daily Return:{avg_daily_return:.4f}")
```

```python
cumulative_returns = (1 + stock_data['Daily Return']).cumprod() - 1
print(f"Cumulative Returns: {cumulative_returns.iloc[-1]:.4f}")
```
